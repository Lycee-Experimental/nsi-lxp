2) Testez le programme suivant.

```python
x = 10 
y = 0

if x < 0:
    y = -20
elif x < 10:
    y = 21
elif 10 < x < 12:
    y = 35
elif x >= 12:
    y = 37
print(y)
```

**Analyse des résultats :**

Le cas 10 n'est pas répertorié dans les différentes conditions, la valeur de y ne change donc pas

Programme corrigé :

```python
x = 10 
y = 0

if x < 0:
    y = -20
elif x < 10:
    y = 21
elif 10 <= x < 12:
    y = 35
elif x >= 12:
    y = 37
print(y)
```

3) Enoncé du problème : 

- si x = 2 ou y = 5, alors z = 35
- si x = 2 et y = 5, alors z = 15
- sinon z = 20

Tester le programme ci-dessous, puis changer les valeurs de x,y par 2,8 puis 2,5

```python
x,y = 4,5

if x == 2 or y == 5:
    z = 35
elif x == 2 and y == 5:
    z = 15
else: 
    z = 20
print(x,y,z)
```

```python
x,y = 2,8

if x == 2 or y == 5:
    z = 35
elif x == 2 and y == 5:
    z = 15
else: 
    z = 20
print(x,y,z)
```

```python
x,y = 2,5

if x == 2 or y == 5:
    z = 35
elif x == 2 and y == 5:
    z = 15
else: 
    z = 20
print(x,y,z)
```

**Analyse des résultats :**

Dans le dernier cas, le programme s'arrête au *if* car les conditions sont bonnes, alors qu'elles sont aussi bonne dans le premier *elif*, il faudrait un ou exclusif. On devrait avoir comme résultat : 2 5 15

Code corrigé :

```python
x,y = 2,5

if x == 2 and y == 5:
    z = 15
elif x == 2 or y == 5:
    z = 35
else: 
    z = 20
print(x,y,z)
```
