# Recherche naïve

```python
def recherche(tableau, element):
    for valeur in tableau:
        if valeur == element:
            return True
    return False
```

# Recherche dans un tableau trié

```python
def recherche_ordonnee(tableau, element):
    i = 0
    N = len(tableau)
    while tableau[i] <= element and i < N:
        if tableau[i] == element:
            return True
        i = i + 1
    return False
```

# Recherche dichotomique itérative

```python
def recherche_dichotomique(tableau, element):
    g = 0
    d = len(tableau) - 1
    while g <= d and tableau[g] <= element <= tableau[d]:
        m = (g + d) // 2
        if tableau[m] == element:
            return True
        elif tableau[m] < element:
            g = m + 1
        else:
            d = m - 1
    # La recherche n'a rien donné
    return False
```

# Recherche dichotomique récursive

```python
def recherche_dichotomique_rec(tableau,element, g, d):
    if g > d:
        # La "largeur" de la zone de recherche est négative: 
        # La recherche a été infructueuse
        return False
    else:
        # On sait que g <= d: la "largeur" de la zone est positive ou nulle
        # (nulle = il ne reste plus qu'un seul élément)
        m = (g + d) // 2
        if tableau[m] == element:
            return True
        elif tableau[m] < element :
            return recherche_dichotomique_rec(tableau, element, m + 1, d)
        else:
            return recherche_dichotomique_rec(tableau, element, g, m - 1)

def recherche_dichotomique_rec_Final(tableau, element):
    return recherche_dichotomique_rec(tableau, element, 0, len(tableau) - 1)
```

# Complexités

```python
from timeit import default_timer as timer
import matplotlib.pyplot as plt
from random import *

def T(n):
    t=[]
    for i in range(n):
        t.append(randint(0,n+1))
    t=sorted(t)
    return t

N=100
t=T(N)

def temps1(n):
    debut=timer()
    recherche_ordonnee(t[:n],max(t[:n]))
    fin=timer()
    tmps1=fin-debut
    return tmps1

def temps2(n):
    debut=timer()
    recherche_dichotomique(t[:n],max(t[:n]))
    fin=timer()
    tmps2=fin-debut
    return tmps2

def temps3(n):             
    debut=timer()
    recherche_dichotomique_rec_Final(t[:n],max(t[:n]))
    fin=timer()
    tmps3=fin-debut
    return tmps3

abscisses = [n for n in range(1,100)]
ordonnees1 = list(map(lambda x: temps1(x), abscisses))
ordonnees2 = list(map(lambda x: temps2(x), abscisses))
ordonnees3= list(map(lambda x: temps3(x), abscisses))


plt.plot(abscisses, ordonnees1,'b')
plt.plot(abscisses, ordonnees2,'r')
plt.plot(abscisses, ordonnees3,'g')


plt.show()
plt.close()
```
