```python
def factorielle1(n):
    fact=1
    for i in range(1,n+1):
        fact=fact*i
    return fact
```

```python
factorielle1(4)
```

```python
def factorielle2(n):
    if n==0:
        return 1
    else:
        return n*factorielle2(n-1)
```

```python
factorielle2(4)
```

```python
import matplotlib.pyplot as plt
from timeit import default_timer as timer

N=1000

def temps1(n):
    debut=timer()
    factorielle1(n)
    fin=timer()
    tmps1=fin-debut
    return tmps1

def temps2(n):
    debut=timer()
    factorielle2(n)
    fin=timer()
    tmps2=fin-debut
    return tmps2

abscisses = [n for n in range(0, N)]
ordonnees1 = list(map(lambda x: temps1(x), abscisses))
ordonnees2 = list(map(lambda x: temps2(x), abscisses))


plt.plot(abscisses, ordonnees1,'b')
plt.plot(abscisses, ordonnees2,'r')


plt.show()
```
