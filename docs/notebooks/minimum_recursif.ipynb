```python
def mini1(L):
    mini1=L[0]
    for i in range(len(L)):
        if L[i]<mini1:
            mini1=L[i]
    return mini1
```

```python
mini1([2,4,1,3])
```

```python
def mini2(L):
    if len(L)==1:
        return L[0]
    else:
        if L[0]<mini2(L[1:]):
            return L[0]
        else:
            return mini2(L[1:])
```

```python
mini2([2,4,5,3,1,7])
```

```python
from timeit import default_timer as timer
import matplotlib.pyplot as plt
from random import *

N=20

def temps1(L):
    debut=timer()
    mini1(L)
    fin=timer()
    tmps1=fin-debut
    return tmps1

def temps2(L):
    debut=timer()
    mini2(L)
    fin=timer()
    tmps2=fin-debut
    return tmps2

listes = [[randint(0,n) for i in range(n)] for n in range(1,N)]
ordonnees1 = list(map(lambda x: temps1(x), listes))
ordonnees2 = list(map(lambda x: temps2(x), listes))
abscisses = [n for n in range(1, N)]


plt.plot(abscisses, ordonnees1,'b')
plt.plot(abscisses, ordonnees2,'r')


plt.show()
```
