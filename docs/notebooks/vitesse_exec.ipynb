```python
def puissance1(n):
    p=1
    for i in range(n):
        p=2*p
    return p
```

```python
def puissance2(n):
    if n==0:
        return 1
    else:
        return 2*puissance2(n-1)
```

```python
from timeit import default_timer as timer

N=100

debut=timer()
print(puissance1(N))
fin=timer()
print(f"temps pour calculer 2^{N} en version itérative : {fin-debut}")

debut=timer()
print(puissance2(N))
fin=timer()
print(f"temps pour calculer 2^{N} en version récursive : {fin-debut}")
```

# Visualisation graphique

```python
import matplotlib.pyplot as plt

def temps1(n):
    debut=timer()
    puissance1(n)
    fin=timer()
    tmps1=fin-debut
    return tmps1

def temps2(n):
    debut=timer()
    puissance2(n)
    fin=timer()
    tmps2=fin-debut
    return tmps2

abscisses = [n for n in range(0, 1000)]
ordonnees1 = list(map(lambda x: temps1(x), abscisses))
ordonnees2 = list(map(lambda x: temps2(x), abscisses))


plt.plot(abscisses, ordonnees1,'b')
plt.plot(abscisses, ordonnees2,'r')


plt.show()
```
